Zuerst hinzufügen in dem Modul smac/runhistory/runhistory.py zu der Klasse RunHistory das Attribut get_all_configs_combined() hinzufügen, da dieses aufgerufen wird in configurator_footprint():

```python
# Nicht ausführbar, nur zum Kopieren hier!!!

class RunHistory(object):

    """Add attribut, needed for BA CAVE version."""

    def get_all_configs_combined(self):
        """Return all configurations in correct order of the RunHistory object

        Returns
        -------
            parameter configurations: list
        """
        order = []
        for index in self.ids_config.keys():
            order.append(self.ids_config[index])
        return order
```

## Create an cave Object

Erstelle das CAVE Object anhand der smac Ergebnisse aus Data_smac. Dabei sind die Ergebnisse der zu visualisierenden Hyperparameter im angegebenen Pfad folders  in der runhistory.json (&rarr; enthält die Informationen über Einstellungen und ob random oder local gesucht wurde).
Aber hier gilt einfach: hier wird das CAVE Objekt erstellt, welches dann alle Informationen enthält zu den mit smac untersuchten Hyperparametern.

```python
from cave.cavefacade import CAVE

cave = CAVE(folders=["./small_example/Data_smac/smac3-output_2019-05-21_10:16:34_055272/run_1/"],
            output_dir="./small_example/Data_smac/CAVE_small_example_Jupyter/",
            ta_exec_dir=["./small_example/Data_smac/"],
            file_format='SMAC3',
            verbose_level='DEBUG',
           )
```

## Call the method

Aufrufen von configurator_footprint(). Hier wird dann ausgegeben, wo die verschiedenen Hyperparameter in einem 2D Koordinatensystem liegen. Weiße Punkte stehen für die random Hyperparameter, weiße Punkte mit Kreuz in der Mitte stehen für die localen Hyperparameter.

Als erstes wird hier eine Einteilung von Runhistory gemacht. Es werden erst die random Konfgurationen hinzugefügt, danach die localen Konfiguratinen. Ahand der Runhistory wird die conf_matrix erstellt, welche für m Hyperparameter mit n Dimensionen dann eine mxn Matrix ist (in diesem Beispiel, 10 Hyperparameter mit 3 kontinuirlichen Dimensionen).
Aus der conf_matrix wird die dists matrix erstellt, welche eine symmetrische Distanzmatrix ist und die Information von jeder Konfiguration zu allen anderen enthält.

### Bengio Method

Meine Implementation von dem Paper. Hier wird die MDS methode mit der Bengio Methode ausgeführt.
Es wird erst die Matrix mit den Distanzen zwischen den random Konfiguratine runterprojeziert. Danach werden die localen Konfigurationen anhand der extend Methode runterprojeziert.

```python
cave.configurator_footprint(reduction_method="classic")
```

### Smacof Method

Zum Vergleich die SMACOF Methode, welche in dem sklearn package implementiert ist.

```python
cave.configurator_footprint(reduction_method="smacof")
```

Zum Vergleich dient der Costvalue. Dieser wird berechnet indem die Differenz zweier Konfigurationen in der hohen und niedrigen Dimension berechnet wird. Kann in der debug.log eingesehen werden, bzw im jupyter notebook in der Ausgabe.
